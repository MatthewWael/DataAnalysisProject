```python
df = pd.read_csv('Influence Studying With Music on Productivity.csv')
df.drop(1, inplace=True)
df.drop('Email Address', axis=1, inplace=True)
df.loc[12, 'What type of music do you listen to while studying?'] = "Hip Hop/Pop"
df.loc[52, 'What type of music do you listen to while studying?'] = "Lofi Music"
for x in df.index:
    music_type = df.loc[x, "What type of music do you listen to while studying?"]
    if music_type in ['Anything', 'Random', 'All of the above']:
        df.loc[x, "What type of music do you listen to while studying?"] = 'All Types'


df.head(55)
```

```python
# to get the mean of "the overall impact of studying with music on your productivity"


Median= df[ 'How would you rate the overall impact of studying with music on your productivity?'].median()
from scipy import stats
Mode= stats.mode(df[ 'How would you rate the overall impact of studying with music on your productivity?'])

print("Median: ", Median)
print("Mode: ", Mode)
```

```python
from matplotlib import pyplot as plt

# Extracting the value counts of each rating
rating_counts = df['How would you rate the overall impact of studying with music on your productivity?'].value_counts()

# Creating labels and data for the pie chart
ratings = rating_counts.index.astype(str)  # Convert index labels to strings
counts = rating_counts.values

# Creating plot
fig, ax = plt.subplots()
ax.pie(counts, labels=ratings, autopct='%1.1f%%', startangle=90)
ax.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle

# Adding title
plt.title('Overall Impact of Studying with Music on Productivity')

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Sample data (replace with your actual data)
data = {
    "Response": ["Yes, it enhances my focus.", "No, it distracts me.", "I haven't tried studying with music before.", "It doesn't enhance nor distract me", "Yes, it enhances my focus."],
}

# Convert data to DataFrame
df = pd.DataFrame(data)

# Count occurrences of each response
response_counts = df['Response'].value_counts()

# Set figure size
plt.figure(figsize=(8, 8))

# Plot pie chart
plt.pie(response_counts, labels=response_counts.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Set3.colors)

# Add title
plt.title('Responses to "Do you find that studying with music helps you concentrate better?"')

# Equal aspect ratio ensures that pie is drawn as a circle
plt.axis('equal')

# Show plot
plt.tight_layout()
plt.show()
```
